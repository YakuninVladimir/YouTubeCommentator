```python
import pandas as pd
from transformers import pipeline, AutoTokenizer
import torch

MODEL_NAME = "blanchefort/rubert-base-cased-sentiment"  
TOP_COMMENTS_N = 30  

print("–®–∞–≥ 1/4: –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö...")
comments['combined_weight'] = comments['likes'] + comments['replies'] * 10

print("–®–∞–≥ 2/4: –û—Ç–±–æ—Ä —Ç–æ–ø–æ–≤—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤...")
top_comments = (
    comments.sort_values(['video_id', 'combined_weight'], ascending=[True, False])
    .groupby('video_id')
    .head(TOP_COMMENTS_N)
    .reset_index(drop=True)
)

print("–®–∞–≥ 3/4: –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–æ–¥–µ–ª–∏...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
classifier = pipeline(
    "sentiment-analysis",
    model=MODEL_NAME,
    tokenizer=tokenizer,
    device=0 if torch.cuda.is_available() else -1
)

def analyze_sentiment_batch(texts):
    results = classifier(texts, truncation=True, max_length=512, batch_size=BATCH_SIZE)
    return [result['label'] for result in results]

print("–®–∞–≥ 4/4: –ê–Ω–∞–ª–∏–∑ —Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç–∏...")
texts = top_comments['comment_text'].astype(str).tolist()
sentiments = []

for i in tqdm(range(0, len(texts), BATCH_SIZE), desc="–û–±—Ä–∞–±–æ—Ç–∫–∞"):
    batch = texts[i:i+BATCH_SIZE]
    try:
        sentiments.extend(analyze_sentiment_batch(batch))
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –≤ –±–∞—Ç—á–µ {i//BATCH_SIZE}: {str(e)}")
        sentiments.extend(['NEUTRAL']*len(batch))

top_comments['sentiment'] = sentiments

top_positive = top_comments[top_comments['sentiment'] == 'POSITIVE']
top_negative = top_comments[top_comments['sentiment'] == 'NEGATIVE']

print("–°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤...")
top_comments.to_csv('top30_comments_with_sentiment.csv', index=False)
top_positive.to_csv('top30_positive_comments.csv', index=False)
top_negative.to_csv('top30_negative_comments.csv', index=False)

print(f"""
–ê–Ω–∞–ª–∏–∑ –∑–∞–≤–µ—Ä—à–µ–Ω —É—Å–ø–µ—à–Ω–æ!
–í—Å–µ–≥–æ –æ–±—Ä–∞–±–æ—Ç–∞–Ω–æ –≤–∏–¥–µ–æ: {top_comments['video_id'].nunique()}
–¢–æ–ø–æ–≤—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤: {len(top_comments)}
–ò–∑ –Ω–∏—Ö:
  - –ü–æ–∑–∏—Ç–∏–≤–Ω—ã—Ö: {len(top_positive)}
  - –ù–µ–≥–∞—Ç–∏–≤–Ω—ã—Ö: {len(top_negative)}
""")
```

```python
top_comments = pd.read_csv('top30_comments_with_sentiment.csv')
videos = pd.read_csv('videos.csv')
```

```python
videos['descriptions'] = videos['descriptions'].fillna('')
```

```python
def truncate_description(desc, max_words=100):
    if pd.isna(desc):
        return ""
    words = str(desc).split()
    if len(words) > max_words:
        return ' '.join(words[:max_words]) + "..."
    return desc

videos['short_description'] = videos['descriptions'].apply(truncate_description)
```

```python

enriched_comments = pd.merge(
    top_comments,
    videos[['video_id', 'title', 'channel_title', 'tags', 'short_description']],
    on='video_id',
    how='left'
)


enriched_comments['enriched_text'] = (
    "sentiment " + enriched_comments['sentiment'] + "\n" + 
    "video: " + enriched_comments['title'] + "\n" +
    "channel: " + enriched_comments['channel_title'] + "\n" +
    "fags: " + enriched_comments['tags'].fillna('').str.replace('|', ', ') + "\n" +
    "description: " + enriched_comments['short_description'].fillna('') + "\n" +
    "comment:\n" + enriched_comments['comment_text']
)


enriched_comments.to_csv('final_enriched_comments.csv', index=False)

print(f"–ì–æ—Ç–æ–≤–æ! –û–±–æ–≥–∞—â–µ–Ω–æ {len(enriched_comments)} –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤")
print("–ü—Ä–∏–º–µ—Ä –æ–±–æ–≥–∞—â–µ–Ω–Ω–æ–≥–æ —Ç–µ–∫—Å—Ç–∞:")
print(enriched_comments['enriched_text'].iloc[0][:1000] + "...")
```

```python
import pandas as pd
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments
from transformers import DataCollatorForLanguageModeling

from datasets import Dataset
import torch


df = pd.read_csv('final_enriched_comments.csv')
texts = df['enriched_text'].tolist()[:10000]


dataset = Dataset.from_dict({'text': texts})


tokenizer = GPT2Tokenizer.from_pretrained("openai-community/gpt2-large")
tokenizer.pad_token = tokenizer.eos_token  


data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False  
)

def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, max_length=512, padding="max_length", return_tensors="pt")

tokenized_dataset = dataset.map(tokenize_function, batched=True)
```

```python
from transformers import AutoModelForCausalLM, GPT2LMHeadModel


model = GPT2LMHeadModel.from_pretrained("openai-community/gpt2-large")

for param in model.base_model.parameters():
    param.requires_grad = False  
    
for param in model.transformer.h[-4:].parameters():  
    param.requires_grad = True
```

```python
training_args = TrainingArguments(
    output_dir="./gpt2_finetuned_extra",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=1, 
    save_steps=10_000,
    save_total_limit=2,
    prediction_loss_only=True,
    learning_rate=5e-5,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=100,
    fp16=torch.cuda.is_available(),
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,  
)
```

```python
print("–ù–∞—á–∏–Ω–∞–µ–º –¥–æ–æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏...")
trainer.train()
print("–î–æ–æ–±—É—á–µ–Ω–∏–µ –∑–∞–≤–µ—Ä—à–µ–Ω–æ!")

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
model.save_pretrained("./gpt2_finetuned_extra")
tokenizer.save_pretrained("./gpt2_finetuned_extra")
```
